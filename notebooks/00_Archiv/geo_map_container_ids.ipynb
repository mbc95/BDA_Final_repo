```python
#geo_df_2 = geo_df.groupby(['laengengrad'])['container_id'].apply(lambda x: ''.join(str(x))).reset_index()
geo_df['container_id'] = geo_df['container_id'].astype(str)
geo_df_2 = geo_df.groupby(['laengengrad', 'breitengrad'], as_index = False).agg({'container_id': '+'.join})
```

```python
output_file("../data/modeling/plots/bokeh_v2.html")

map_options = GMapOptions(lat = 50.13, lng = 8.73, map_type = "roadmap", zoom = 12)
bokeh_plot = gmap(AUTH_KEY, map_options, title = "Container locations")

bokeh_plot.circle(x = "breitengrad", y = "laengengrad", size = 10, fill_alpha = 0.8, source = geo_df_2)

source = ColumnDataSource(dict(x = geo_df_2['breitengrad'], y = geo_df_2['laengengrad'], text = geo_df_2['container_id']))

glyph = Text(x="x", y="y", text="text", angle=0, text_color="#000000", text_font_size="10px")
bokeh_plot.add_glyph(source, glyph)

show(bokeh_plot)
```
