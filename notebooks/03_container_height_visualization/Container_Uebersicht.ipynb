```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
df_garbage = pd.read_csv("../data/preprocessed/cleaned_data_v3.csv")
```

```python
# set datum as index
height = df_garbage[["container_id", "height_in_cm", "datum"]]
height["datum"] = pd.to_datetime(height['datum'])
height = height.set_index("datum")
```

```python
brauchbar = [1,4,5,7,10,15,19,22,24,29,34,35,36,45,47,48,49,52,54,56,58,66,68,72,76]
bearbeiten = [3,6,9,13,14,20,21,23,26,28,30,31,32,33,37,38,39,40,41,42,43,46,50,53,59,61,62,63,64,65,67,69,73,75,77]
unbrauchbar = [2,8,11,12,16,17,18,25,27,55,57,60,70,71,74]
```

# Brauchbar

```python

for i in brauchbar:
    plt.figure(figsize=(12,3))
    plt.plot(height["height_in_cm"].loc[(height["container_id"]==i)])
    plt.title("Container "+ str(i))
    plt.ylim(0,150)
    plt.show()
```

# Zu bearbeiten

```python
for i in bearbeiten:
    plt.figure(figsize=(12,3))
    plt.plot(height["height_in_cm"].loc[(height["container_id"]==i)])
    plt.title("Container "+ str(i))
    plt.ylim(0,150)
    plt.show()
```

```python
for i in unbrauchbar:
    plt.figure(figsize=(12,3))
    plt.plot(height["height_in_cm"].loc[(height["container_id"]==i)])
    plt.title("Container "+ str(i))
    plt.ylim(0,150)
    plt.show()
```
